**Matplotlib - Pie Chart, Bar Plot, Sine Wave**


```python
#1. Import library
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#2. Creating datas
Movies = ['JL','Avengers','Batman','Ironman','Mask']
Percentage = [23,30,21,19,7]
```

```python
#3. Pie chart
plt.pie(Percentage, labels=Movies)
```

```python
#4. Bar Plot
plt.bar(Movies, Percentage)
```

```python
#Sin wave Plot
#5. importing the libraries
import numpy as np
```

```python
#6. Range Function
time = np.arange(0, 10, 0.05)
```

```python
#7. Sin Operation
amplitude = np.sin(time)
plt.plot(time, amplitude)
```

```python
#Contour Plot
#8.1.Creating datas
data1 = np.linspace(-5, 5, 100)
data2 = np.linspace(-5, 5, 100)
```

```python
#8.2. Creating it as 2D
[x, y] = np.meshgrid(data1, data2)
```

```python
#8.3. Creating 3rd data
z = x**2 + y**2
```

```python
#8.4. Plotting contour
figure, ax = plt.subplots(1,1)
ax.contourf(x, y, z)
ax.set_title('Contour Plot(Filled)')
ax.set_xlabel('x')
ax.set_ylabel('y')
plt.show()
```

```python
#9. Box Plot
#9.1. Creating datas
data1 = np.random.normal(100, 10, 100)
data2 = np.random.normal(80, 30, 100)
data3 = np.random.normal(60, 50, 100)
```

```python
#9.2. Combining datas
data = [data1, data2, data3]
```

```python
#9.3. Plotting box plot
figure = plt.figure(figsize=(8, 5))
ax = figure.add_axes([0, 0, 1, 1])
ax.boxplot(data)
plt.show()
```

```python
#10. scatter Plot
# Import libraries
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt


# Creating dataset
z = np.random.randint(100, size =(50))
x = np.random.randint(80, size =(50))
y = np.random.randint(60, size =(50))

# Creating figure
fig = plt.figure(figsize = (10, 7))
ax = plt.axes(projection ="3d")

# Creating plot
ax.scatter3D(x, y, z, color = "green")
plt.title("simple 3D scatter plot")

# show plot
plt.show()
```

```python
#11. 3D Surface plot
#11.1. Creating datas
x = np.outer(np.linspace(-5,5,25), np.ones(25))
y = x.copy().T
z = (np.sin(x**2) + np.cos(y**2))
```

```python
#11.2. Plotting 3D surface
figure = plt.figure(figsize = (10, 7))
ax = plt.axes(projection = '3d')
ax.plot_surface(x, y, z)
plt.show()
```
